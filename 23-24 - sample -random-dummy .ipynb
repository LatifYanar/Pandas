```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.sample()    # at every run it bribgs a different random row
```

```python
ufo.sample(n=5)   # if you need more rows, dedicate it
```

```python
ufo.sample(n=3, random_state=23)   # if you want same rows everytime, use ramdom_state parameter
```

# creating dummy variables

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
train['sex_male'] = train.Sex.map({'female':0, 'male':1})  # creating a dummy column using Sex column.
```

```python
train.head()
```

```python
pd.get_dummies(train.Sex) # we see the dummy variables created usind/on the sex column4
```

```python
pd.get_dummies(train.Sex).iloc[:,1:]   # dropped the first column
```

```python
pd.get_dummies(train.Sex, prefix='Sex').iloc[:,1:] # added the word'sex' to our column
```

```python
train.Embarked
```

```python
train.Embarked.value_counts()
```

```python
pd.get_dummies(train.Embarked, prefix='Embraked')
```

```python
pd.get_dummies(train.Embarked, prefix='Embraked').iloc[:,1:]  # if you have k variables, you may have k-1 different var?
  # we dropped column c.  if q and s is 0 it is c it means from this frame
```

```python
embarked_dummies = pd.get_dummies(train.Embarked, prefix='Embraked').iloc[:,1:] # saved it 
```

```python
# use another method concat
train = pd.concat([train, embarked_dummies], axis=1)
```
