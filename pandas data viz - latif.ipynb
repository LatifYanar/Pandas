## .plot.hexbin()

```python
# create a new dataframe

df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])

# we want 1000 random numbers for two columns, if we use only(1000) it makes error. shouls be 1000,2 for two columns
```

```python
df.head()
```

```python
df.plot.hexbin('a','b')  # hexagons are very small
```

```python
df.plot.hexbin('a','b', gridsize=30)
```

```python
df.plot.hexbin('a','b', gridsize=30, cmap='rainbow')
```
