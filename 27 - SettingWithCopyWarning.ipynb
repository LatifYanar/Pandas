```python
# instead of 'not rated's we will overwrite 'Nan'. it is special value prom numpy. and we will import it first

import numpy as np
```

```python
movies[movies.content_rating=='NOT RATED'].content_rating = np.nan   # assigning nan instead of not rated. oppps
```

```python
# this is a warning not an error. it still works. but lets see why this happens.:

movies.content_rating.isnull().sum()     # it is still 3. it did not work. if it worked it should be 65 not 3
```

```python
# in the worning above: Try using .loc[row_indexer,col_indexer] = value instead... it says

movies.loc[movies.content_rating == 'NOT RATED', 'content_rating'] = np.nan
```

```python
movies.content_rating.isnull().sum()   # it worked. use ""loc""   and no warnign
```

###Â what happend? pandas don't guarantee to show a copy of a view of the dataframe. 
* here at the left side we want to see a part of data and at the right side we want to assign something to that.
* and this makes a warning. 
* and we saw that it did not worked. it did not assign. 
* to avoid this to see apart of data use loc[]. it guarantee to bring a part of dataframe.

```python
top_movies = movies.loc[movies.star_rating >= 9, : ]
```

```python
top_movies
```

```python
# if we realize a value here is not rignt and we want to correct it: row 0 duration should be 150. for example.

top_movies.loc[0, 'duration'] = 150
```

```python
top_movies.loc[0, 'duration']  # it changed but warned although
```

### why it generated a warning?
* here because pandas can not be sure that top_movies is a copy of a view of movies. 
* and it is confsed if you want to modify one or two things.
* solution: problem is from the line we created top_movies.   -    we should use copy() method.

```python
top_movies = movies.loc[movies.star_rating >= 9, : ].copy() # by this way pandas is sure that it is a copy
```

```python
top_movies.loc[0, 'duration'] = 150   # it did not warned. 
```
