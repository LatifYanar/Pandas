# Selecting multiple rows and columns

## .loc[  :  ,  :  ]       #  includes both sides

### for labels. (iloc :  for integers)

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
# loc is for selecting by lable

ufo.loc[0,:]    # row 0 column all
```

```python
ufo.loc[0:5,'City']    
```

```python
ufo.loc[0:5,'City':'State']      # 0 and 5 included, City and State included
```

```python
ufo.loc[[0,1,2,5],['City','State']]    # usese arrays inside
```

```python
ufo.loc[0:5]     # it is not mentioned and pandas understands that you want to see all columns 
```

```python
ufo.loc[:, :].head()    # all rows and all columns 
```

```python
ufo.loc[ufo.City == 'Oakland' , :]     # rows : the one have oakland as City, columns: all
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.shape
```

```python
ufo.head()
```

```python
ufo.drop('City', axis=1)   # city is gone
```

```python
ufo.head()   # but not gone.  because we did not use inplace
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.drop('City', axis =1, inplace= True)   # didn't print anything, it show there is something done inside
```

```python
ufo.head()    # gone really
```

```python
ufo.fillna(method='bfill').tail()    # fills the gap using the next valid value
```

```python
ufo.fillna(method='ffill').tail() # uses the last valid value to fill forward gap
```

### here continent column is string(object)  and it uses more memory than integer. 

### if we use a system of numbers from 0 to 1 each of them representing a continent, it will use less memory. 
### we will again list strings but just for one time

### this is called 'category' type

```python
# how to do it?

drinks['continent'] = drinks.continent.astype('category')   # that is it :)
```

```python
drinks.dtypes    # continent is a categoty
```

```python
drinks.continent.head() # still the same names. but below it shows us category. a list of continents
```

```python
drinks.memory_usage(deep=True)   # continent 12332 to 744
```

```python
drinks.continent.cat.codes  # a category. ...    method. str. .. gibi.  showing us the real continents :)
```

```python
# lets do it for country 
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks['country'] = drinks.country.astype('category')
```

```python
# check the memory

drinks.memory_usage(deep=True)     # from 17800 to 18000. it got larger. there is a lot of country and the set is big
                                    # but there was only 5 continents
```

```python
drinks.country.cat.categories   # there are 193 categories
```

## in what way can it (category) be used 
## very good !! and helpful



```python
# create a new data frame

df = pd.DataFrame({'ID':[100,101,102,103], 'quality':['good', 'very good', 'good', 'exellent']})
```

```python
df
```

```python
df.sort_values('quality')   # when we call the quality serie it sorts it in alphabetical order
```

```python
# what if I want it to be ordered 'logically'

df['quality'] = df.quality.astype('category', categories=['good','very good', 'exellent'], ordered=True)
```

```python
df.sort_values('quality')    # this time it sorted according to our order.
```

```python
# a better usage

df.quality     # it is ordered
```

```python
df.loc[df.quality > 'good']    # I can see the students better than good
```
